```python
import pandas as pd
import numpy as np
```

There is an invalid value coded as "xx" in the height(cm) column. convert invalid values to "NaN"

```python
df['height(cm)'] = pd.to_numeric(df['height(cm)'], errors='coerce')
print(df)
```

 split the 'age_sex' column into two separate columns.

```python
df[['age', 'sex']] = df['age_sex'].str.split('_', expand=True)
print(df)
```

```python
desired_order = ['age', 'sex', 'height(cm)', 'weight(kg)', ...]
df = df.reindex(columns=desired_order)

print(df)
```
